```python
import pandas as pd
```

reading file auto.csv and creating columns datetime and user, separated by \t
df for print

```python
df = pd.read_csv('feed-views.log', names=['datetime', 'user'], sep='\t')
df
```

converting the datetime to the datetime64[ns] Dtype<br>
extracting the year, month, day, hour, minute, and second from the datetime

```python
df['datetime'] = pd.to_datetime(df['datetime'])
df['year'] = df['datetime'].dt.year
df['month'] = df['datetime'].dt.month
df['day'] = df['datetime'].dt.day
df['hour'] = df['datetime'].dt.hour
df['minute'] = df['datetime'].dt.minute
df['second'] = df['datetime'].dt.second
df
```

creating list "label" and "intervals"
0.00 – 03.59 night, 04.00 – 06.59 early morning, 07.00 – 10.59 morning, 11.00
– 16.59 afternoon, 17.00 – 19.59 early evening, 20.00 – 23.59 

bins and labels are bound
labels must be one fewer than the number of bin edges 
if 0.00 – 03.59 night we return label 'night' and etc.
include_lowest=True shows whether the left interval be in an inclusion 
right=False shows whether the right interval be in an inclusion

assigning the column user as the index

```python
day_times = ['night', 'early morning', 'morning', 'afternoon', 'early evening', 'evening']
intervals = [0, 4, 7, 11, 17, 20, 24]
df['daytime'] = pd.cut(df['hour'], labels=day_times, bins=intervals, include_lowest=True, right=False)
df.set_index('user', inplace=True)
df
```

calculating the number of elements in df 

```python
df.count()
```

value_counts() calculates the number of elements in each time of day category using the method


```python
df.value_counts('daytime')
```
